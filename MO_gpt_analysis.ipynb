```python
import pandas as pd
import glob
import json
import os
```

```python
# 

dict_of_logs = {}
for name in glob.glob('logs/*'):
    with open(name, "rb") as f :
        dict_resp = json.load(f)
        dict_resp['resp'] = dict_resp['choices'][0]['message']['content']
        dict_of_logs[name] = dict_resp

df = pd.DataFrame.from_dict(dict_of_logs, orient='index')
df = df[df['resp'] != ""]
df["gen_label"]= [1 if resp.find("1")>-1 else 0 for resp in df["resp"]]
df['agree'] = df['label']==df['gen_label']

df.to_csv("chat_mo_results.csv")

df.groupby(['model','prompt_type']).agg(
     tot_resp = ('agree','count'),
     num_correct = ('agree','sum'),
     ).reset_index()
```
